```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 2 * np.pi, 50)
plt.plot(x, np.sin(x))
plt.show() # Show the graph.
```

```python
plt.plot(x, np.sin(x),
         x, np.sin(2 * x))
plt.show()
```

```python
plt.plot(x, np.sin(x), 'r-o',
         x, np.cos(x), 'g--')
plt.show()
```

```python
plt.subplot(2, 1, 1) # (row, column, active area)
plt.plot(x, np.sin(x), 'r')
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x), 'g')
plt.show()
```

```python
y = np.sin(x)
plt.plot(x, y)
plt.show()
```

```python
x = np.random.rand(1000)
y = np.random.rand(1000)
size = np.random.rand(1000) * 50
color = np.random.rand(1000)
plt.scatter (x, y, size, color)
plt.colorbar()
plt.show()
```

```python
x = np.linspace(0, 2 * np.pi, 50)
plt.plot(x, np.sin(x), 'r-x', label='Sin(x)')
plt.plot(x, np.cos(x), 'g-^', label='Cos(x)')
plt.legend() # Display the legend.
plt.xlabel('Rads') # Add a label to the x-axis.
plt.ylabel('Amplitude') # Add a label to the y-axis.
plt.title('Sin and Cos Waves') # Add a graph title.
plt.show()
```

```python
i = 0
j = 0
t = 1
z = np.empty([2001,2001])
for x in np.arange(-10,10,.01):
    i = i + 1
    j = 0
    for y in np.arange(-10,10,.01):
        j = j + 1
        z[i][j] = x*x + y*y
```

```python
def f1(x,y):
    return (x*x+y*y)
x = np.linspace(-10, 10, 2000)
y = np.linspace(-10, 10, 2000)
X, Y = np.meshgrid(x, y)
Z = f1(X, Y)
```

```python
Z
```

```python
fig1 = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='hot')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
ax.view_init(45, 0)
```
