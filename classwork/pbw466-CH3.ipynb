```python
import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
def monteCarlo(num): 
    count = 0
    for x in range(0, num):
        x=((np.random.rand())*2)-1
        y=((np.random.rand())*2)-1
        if (((x*x)+(y*y))<1):
            count+=1
    return (4.0 * count / num)
```

```python
x=[]
y=[]
z=[]

for i in range(0, 100):
    calcPi=monteCarlo(1000)
    absError=abs(np.pi - calcPi)
    relError=abs((np.pi - calcPi)/np.pi)
    x.append(i)
    y.append(absError)
    z.append(relError)
plt.subplot(2,1,1)
plt.plot(x,y)
plt.subplot(2,1,2)
plt.plot(x,z)
plt.show()
```
