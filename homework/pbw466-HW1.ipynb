```python
import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
def monteCarlo(num): 
    count = 0
    for x in range(0, num):
        x=((np.random.rand())*2)-1
        y=((np.random.rand())*2)-1
        if (((x*x)+(y*y))<1):
            count+=1
    return (4.0 * count / num)
```

```python
x=[]
y=[]
for i in range(10, 10000, 20):
    error=abs(math.pi - monteCarlo(i))
    x.append(i)
    y.append(error)
plt.plot(x,y)
plt.show()
```
